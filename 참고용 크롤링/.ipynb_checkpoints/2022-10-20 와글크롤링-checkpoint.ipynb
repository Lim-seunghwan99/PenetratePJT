```python
url = 'https://www.changwon.ac.kr/portal/main.do'
wd = webdriver.Chrome('C:/Users/cwg501/Desktop/chromedriver.exe')
#driver.get(url)
wd.get(url)
results = []
time.sleep(1)
#driver.find_element(By.XPATH, '//*[@id="container"]/div[2]/div[3]/div/div/div[1]/a').click()
wd.find_element(By.XPATH, '//*[@id="container"]/div[2]/div[3]/div/div/div[1]/a').click()

for page in range(1, 10) :
    try :
        wd.execute_script('goPaging(%d)'%page)
        time.sleep(1)
        html = wd.page_source
        bs = BeautifulSoup(html, 'html.parser')
        tr_tags = bs.select('tr')    
        for i in range(5, 16) :
            num = tr_tags[i].select('td')[0].string
            title = tr_tags[i].select('td')[1].a.string.strip()
            date = tr_tags[i].select('td')[2].string
            view = tr_tags[i].select('td')[3].string
            print(num,title,date,view)
            results.append([num,title,date,view])
    except Exception as e :
        print(e)
```

```python
url = 'https://www.changwon.ac.kr/portal/main.do'
wd = webdriver.Chrome('C:/Users/cwg501/Desktop/chromedriver.exe')
#driver.get(url)
wd.get(url)
results = []
time.sleep(1)
#driver.find_element(By.XPATH, '//*[@id="container"]/div[2]/div[3]/div/div/div[1]/a').click()
wd.find_element(By.XPATH, '//*[@id="container"]/div[2]/div[3]/div/div/div[4]/a/i').click()
#driver.find_element(By.XPATH, '//*[@id="container"]/div[2]/div[3]/div/div/div[4]/a/i').click()
wd.switch_to.window(wd.window_handles[1])
#driver.switch_to.window(wd.window_handles[1])

for page in range(1, 10) :
    try :
        wd.execute_script('goPaging(%d)'%page)
        time.sleep(1)
        html = wd.page_source
        bs = BeautifulSoup(html, 'html.parser')
        tr_tags = bs.select('tr')    
        for i in range(5, 16) :
            num = tr_tags[i].select('td')[0].string
            title = tr_tags[i].select('td')[1].a.string.strip()
            name = tr_tags[i].select('td')[2].string.strip()
            date = tr_tags[i].select('td')[3].string
            view = tr_tags[i].select('td')[4].string
            results.append([num,title,name,date,view])
    except Exception as e :
        print(e)
print(results)
```
